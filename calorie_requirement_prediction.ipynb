```python
cols = ['Gender',
       'Activity_Level', 'Health_Status']
```

```python
for col in cols:
    print(f"\n{'='*40}")
    print(f"{col}")
    print(f"{'='*40}")
    print(new_df[col].unique())
```

```python
for col in cols:
    if col == 'Gender':
        new_df[col] = new_df[col].map({'Male' : 0, 'Female' : 1, 'Other' : 2})
    elif col == 'Activity_Level':
        new_df[col] = new_df[col].map({'Sedentary':0, 'Lightly Active' : 1, 'Moderately Active': 2, 'Very Active': 3, 'Athlete': 4}) 
    elif col == 'Health_Status':
        new_df[col] = new_df[col].map({'Underweight': 0, 'Healthy': 1, 'Overweight': 2, 'Obese' : 3})
  
```

```python
new_df = pd.get_dummies(new_df, columns = ['Diet_Type'], drop_first = True, dtype = int)
```

```python
new_df.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs = new_df.drop(columns=[
    'Daily_Calorie_Requirement', 
    'BMI', 
    'Daily_Calorie_Consumed', 
    'Protein_Intake_g', 
    'Carbohydrate_Intake_g', 
    'Fat_Intake_g'
])
```

```python
target = new_df['Daily_Calorie_Requirement']
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators = 100, random_state = 42)
```

```python
rf_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
y_pred = rf_model.predict(x_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f} kcal")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f} kcal")
print(f"R² Score: {r2:.4f}")
```
