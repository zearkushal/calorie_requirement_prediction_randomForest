```python
new_df = df.drop('Person_ID', axis = 1)
```

```python
new_df.head()
```

```python
new_df.columns
```

```python
cols = ['Gender',
       'Activity_Level', 'Health_Status']
```

```python
cols
```

```python
for col in cols:
    print(new_df[col].describe(include = 'unique'))
```

```python
for col in cols:
    print(f"\n{'='*40}")
    print(f"{col}")
    print(f"{'='*40}")
    print(new_df[col].unique())
```

```python
for col in cols:
    if col == 'Gender':
        new_df[col] = new_df[col].map({'Male' : 0, 'Female' : 1, 'Other' : 2})
    elif col == 'Activity_Level':
        new_df[col] = new_df[col].map({'Sedentary':0, 'Lightly Active' : 1, 'Moderately Active': 2, 'Very Active': 3, 'Athlete': 4}) 
    elif col == 'Health_Status':
        new_df[col] = new_df[col].map({'Underweight': 0, 'Healthy': 1, 'Overweight': 2, 'Obese' : 3})
  
```

```python
new_df[cols]
```

```python
new_df.head()
```

```python
new_df['Diet_Type'].unique()
```

```python
diet_col = ['Keto', 'Vegan', 'High Protein', 'Balanced', 'Mediterranean',
            'Vegetarian']

diet_map = {diet: i for i, diet in enumerate(diet_col)}

new_df['Diet_Type'] = new_df['Diet_Type'].map(diet_map)
```
